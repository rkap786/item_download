```python
# Import some Python modules
# The first section are those built-in to Python;
# the second is third-party modules that must be installed (though they're already
#  installed by default here in Google Colab).
import os
import re
import re
from pathlib import Path

import requests
from bs4 import BeautifulSoup
```

```python
# Fetch the HTML of the source web page...
# Texas
r = requests.get("https://tea.texas.gov/student-assessment/testing/staar/staar-released-test-questions")

# ...and parse it so we can query for the parts we're interested in.
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
# Extract all the elements with an `href` attribute whose value ends in `.pdf`
pdf_links = soup.find_all(href=re.compile("\.pdf$"))

# Loop through these elements and print the text and the href
for pdf_link in pdf_links[:10]:
    print(f"{pdf_link.text:52} | {pdf_link.attrs['href']}")

print(f"\nShowing up to 10 links out of a total of {len(pdf_links)}.")
```

```python
# Select an output directory for the downloads and ensure it exists
output_dir_read = Path("/content/drive/MyDrive/0. Projects - Stanford/Item generation/items read")
output_dir_math = Path("/content/drive/MyDrive/0. Projects - Stanford/Item generation/items math")
output_dir.mkdir(exist_ok=True, parents=True)
```

```python
for pdf_link in pdf_links:
    pdf_href = pdf_link.attrs['href']

    # Construct a filepath from the output_dir and the last section of the href
    #  (i.e. the filename)
    pdf_filepath = output_dir / pdf_href.split('/')[-1]

    # Many links are relative, so add the protocol and the domain to create a valid URI
    if not pdf_href.startswith("http"):
        pdf_href = "https://tea.texas.gov" + pdf_href

    # Download the file :)
    print(f"Downloading {pdf_href} to {pdf_filepath}...")
    with requests.get(pdf_href) as r:
        with pdf_filepath.open("wb") as fh:
            fh.write(r.content)

    # ...profit?
```
